### 3.2 Verifying redundancies in our data

Let's verify the redundancies identified earlier in the project : 

#### 3.2.1 Checking the domestic supply formula

Domestic Supply = Production + imports - exports - changes in stocks (decrease or increase) 

```python
#Let's check the first formula to calculate Domestic Supply :
print(len(data.loc[data.domestic_supply_quantity !=
                   (data.production +data.import_quantity - data.export_quantity - data.stock_variation)]))
print(len(data))
##7775 out of 33515 rows do not abide by that rule

#Let's examine the data for one of the columns that do not respect this rule
veg.loc[(veg["Area"] == "Armenia") & (veg["Item Code"] == 2514) & (veg.Year == 2018)]

#We can see that if we apply the formula, we end up at 88 instead of 87. Maybe this is a rounding issue?

#Testing the rounding issue hypothesis, we check if the absolute value of domestic_supply_quantity - its expected quantity calculated with the formula is less or equal than 2
data.loc[abs(data.domestic_supply_quantity -
             (data.production +data.import_quantity - data.export_quantity - data.stock_variation)) > 2]

#We now have only 110 rows were this is wrong (156 for difference > 1), let's investigate the first one :
data.loc[(data.country_code == 9) & (data.item_code == 2762) & (data.year == 2014)]

#If we calculate the formula by hand, we see that it would work if we changed the sign of stock_variation.
subset = data.loc[abs(data.domestic_supply_quantity -
                      (data.production +data.import_quantity - data.export_quantity - data.stock_variation)) > 2]

subset.loc[abs(data.domestic_supply_quantity -
               (data.production +data.import_quantity - data.export_quantity + data.stock_variation)) > 1]
#This explains all values, we now have 0 differences
```

#### Analyzing our comparison

The question is now what value is more reliable, the domestic supply quantity or the formula? 

When going back into our source files, it seems like all variables used in the formula are "calculated using imputation methodology" whereas the Domestic Supply data is standardized. We are inclined to believe that the **domestic supply data is the right value**.

The only thing we will do here is to invert the sign of the stock variation for the 110 columns where it is not coherent :

```python
data.loc[abs(data.domestic_supply_quantity - 
             (data.production +data.import_quantity - data.export_quantity - data.stock_variation)) > 2,
         "stock_variation"] = -data.stock_variation

data.loc[abs(data.domestic_supply_quantity -
             (data.production +data.import_quantity - data.export_quantity - data.stock_variation)) > 2]
```

All values now respect the formulas, with an accepted error of 2.

Since we will study **global trends**, we assess that correcting this error is **not mandatory** but it could be problematic if we were to study more precise numbers.

#### 3.2.2 Checking the food formula

Food = Domestic Supply - Feed - Seed - Losses - Processing - Other uses - Residuals - Tourist Consumption

Let's do the same process for our food formula :

```python
data.columns

data.loc[data.food != 
         (data.domestic_supply_quantity - data.feed - data.seed - data.losses -
          data.processing - data.other_uses - data.residuals - data.tourist_consumption)]
#We still get 5050 rows out of 33515 that do not respect that formula

#Let's calculate with the absolute value technique used above with an error of 1
data.loc[abs(data.food -
        (data.domestic_supply_quantity - data.feed - data.seed - data.losses -
         data.processing - data.other_uses - data.residuals - data.tourist_consumption)) > 1]
#Only 44 rows do not respect the formulas

#Let's try with an accepted error of 2
data.loc[abs(data.food -
        (data.domestic_supply_quantity - data.feed - data.seed - data.losses -
         data.processing - data.other_uses - data.residuals - data.tourist_consumption)) > 2]
#We get only 1 row which doesn't abide by the formule, let's investigae further
subset = data.loc[abs(data.food -
        (data.domestic_supply_quantity - data.feed - data.seed - data.losses -
         data.processing - data.other_uses - data.residuals - data.tourist_consumption)) > 2]

[subset.food, subset.domestic_supply_quantity - subset.feed - subset.seed - subset.losses - 
 subset.processing - subset.other_uses - subset.residuals - subset.tourist_consumption]

#Seems like there is a mistake in the food amount, it is equal to 0.

#Let's correct this mistake by replacing the food value by the calculated value with the formula : 
data.loc[abs(data.food -
        (data.domestic_supply_quantity - data.feed - data.seed - data.losses - data.processing - data.other_uses -
         data.residuals - data.tourist_consumption)) > 2, "food"] = data.domestic_supply_quantity - data.feed - data.seed - data.losses - data.processing - data.other_uses - data.residuals - data.tourist_consumption

#Verify that all columns abide by the formula :
data.loc[abs(data.food -
        (data.domestic_supply_quantity - data.feed - data.seed - data.losses -
         data.processing - data.other_uses - data.residuals - data.tourist_consumption)) > 2]
```

# 4. Computing new variables

```python
#Creating a subset of our data with only 2018 information
data_2018 = data.loc[data["year"]==2018].copy()

#Creating a subset of our population data with only 2018 and the relevant columns (total population)
pop_2018 = pop[(pop.element_code == 511) & (pop.year_code == 2018)][['Value','country_code']].copy()

#Multiplying our values by 1000 which is the base unit of our population
pop_2018["pop"] = pop_2018["Value"]*1000

#Keeping only relevant columns
pop_2018 = pop_2018[['country_code','pop']]

#Merging relevant information
data_2018 = pd.merge(data_2018,pop_2018, how = "left", on = "country_code")

data_2018.head()
```

### Discrepancies between food and food_supply_kcalcapitaday / kgcapitayr

We have found some data with a food value of 0 but food_supply_kcalcapitaday / kgcapitayr not equal to 0

Since the food data is rounded to the nearest integer but not the food_supply_data, we will assume the food data can be replaced by multiplying the quantity in kgcapitayr by the population and dividing it by 1000000 (since food data is in 1000 tons)

```python
#There are more than 1000 rows concerned
data_2018.loc[(data_2018.food == 0) & (data_2018.food_supply_kcalcapitaday !=0)][["country","item","food", "food_supply_kcalcapitaday","food_supply_quantity_kgcapitayr"]]

#We will first create a subset of our database to verify that our results make sense (ideally the food values should be near 0)
subset = data_2018.loc[(data_2018.food == 0) & (data_2018.food_supply_kcalcapitaday !=0)].copy()

subset["food"] = subset["food_supply_quantity_kgcapitayr"]*subset["pop"]/1000000

#This works for the subset so we will do the same for our data, the maximum value we have calculated is 0.6 tons
subset[["food"]].describe()


data_2018.loc[(data_2018.food == 0), "food"] = data_2018["food_supply_quantity_kgcapitayr"]*data_2018["pop"]/1000000

data_2018.head()
```

There could be a valid argument for replacing ALL food values by this calculated variable, but this would require an indepth analysis of our data source which is not assessed to be necessary since we are interested only in **global trends**.

```python
#Calculating new variables
data_2018["food_supply_kcal"] = data_2018["food_supply_kcalcapitaday"]*data_2018["pop"]*365

#Kgprotein needs to be divided by 1000 in the end to convert g into kgs
data_2018["food_supply_kgprotein"] = data_2018["protein_supply_quantity_gcapitaday"]*data_2018["pop"]*365/1000

#Food is expressed in 1000 tonnes, so we have to multiply our results by 1000 twice to get it in kg
data_2018["food_supply_kg"] = data_2018["food"]*1000*1000

#Adding a where close to prevent division by 0 errors
data_2018["ratio_kcalkg"] = np.where(data_2018["food_supply_kg"] != 0,
                                          data_2018["food_supply_kcal"] / data_2018["food_supply_kg"], float("nan") )

#Same here for the calculation of our protein percentage
data_2018["protein_percentage"] = np.where(data_2018["food_supply_kg"] != 0,
                                          data_2018["food_supply_kgprotein"] / data_2018["food_supply_kg"],float("nan") )
 
data_2018.head()
```

```python
#Let's find the average ratios kcal/kg and protein percentage for each item to fill NAN values
rp_data = data_2018.groupby(["item_code","item"]).mean().reset_index()[["item_code","ratio_kcalkg","protein_percentage"]]

#Creating a table with only the missing rows
missing_data = data_2018.loc[data_2018.ratio_kcalkg.isna()].copy().drop(columns=["ratio_kcalkg","protein_percentage"]).merge(rp_data, how="left", on="item_code")

#Selecting only useful columns
missing_data = missing_data[["country_code","item_code","ratio_kcalkg","protein_percentage"]]

#Renaming the columns to avoid duplication (ratio_kcalkg_x and y) after merging
missing_data.columns = ["country_code","item_code","avg_ratio_kcalkg","avg_protein_percentage"]

#Merging our data_2018 main file with missing data
data_2018 = data_2018.merge(missing_data, how="left", on=["country_code","item_code"])

#Replacing NAN values with average values calculated with the aggregation over item code
data_2018.loc[data_2018.ratio_kcalkg.isna() , "ratio_kcalkg"] = data_2018.avg_ratio_kcalkg 
data_2018.loc[data_2018.protein_percentage.isna() , "protein_percentage"] = data_2018.avg_protein_percentage

#Dropping the now useless columns
data_2018.drop(columns=["avg_ratio_kcalkg","avg_protein_percentage"], inplace=True)

data_2018.head()
```

```python
#Second part of our new variables definition :
data_2018["dom_sup_kcal"] = data_2018["domestic_supply_quantity"]*1000*1000*data_2018["ratio_kcalkg"]

data_2018["dom_sup_kgprot"] = data_2018["domestic_supply_quantity"]*1000*1000*data_2018["protein_percentage"]

data_2018.head()
```

### Calculating if the highest imports of the 25 most exported items come from countries with more than 10% malnourishment


Task 5 edit : After analysis of previous data, we realize that a lot of malnutrition numbers are missing for the most recent year (2018). We will fix that by taking the maximum year_code for each countries where undernutrition population numbers are reported. 

```python
malnutrition = pd.read_csv("Data/malnutrition.csv")

#Selecting only relevant columns
malnutrition = malnutrition[["Area Code (FAO)","Area","Item Code","Item","Year Code","Year","Value"]]

#Renaming the columns to comply with naming standards
malnutrition.columns = ["country_code","country","item_code","item","year_code","year","value"]

#First we select only the undernutrition population values
malnutrition_pop = malnutrition[malnutrition.item_code==210011].copy()
#We coerce the values into string, filling incorrect values with 0
malnutrition_pop["undernut_pop"] = pd.to_numeric(malnutrition_pop.value, errors='coerce').fillna(0, downcast='int')

#Selecting only valid non-null values
malnutrition_pop = malnutrition_pop[(malnutrition_pop.undernut_pop > 0)]

#Selecting the latest year where values were valid
malnutrition_pop = malnutrition_pop.groupby("country").max("year_code").reset_index()

#Merging these information with our main malnutrition dataframe (selecting only the latest year with relevant data)
malnutrition = pd.merge(malnutrition,malnutrition_pop[["country_code","year_code"]],how="inner",on=["country_code","year_code"])

#Selecting only undernutrition percentage values and relevant columns
malnut_pct= malnutrition[malnutrition.item_code == 210041][["country_code","country","year","value"]]

malnut_pct.head()
```

```python
#Turning all convertible values into numeric, and non convertible values into 0
malnut_pct.value = pd.to_numeric(malnut_pct.value, errors='coerce').fillna(0, downcast='int')

#Creating a subset with only countries with more than 10% malnourishment
malnut_10pct = malnut_pct[(malnut_pct.value > 10)][["country_code","value"]]

malnut_10pct.rename(columns={"value": "undernut_pct"},inplace=True)

#Joining our country data with data_2018 to gather all information about the concerned countries
malnut_data = pd.merge(data_2018,malnut_10pct, how = "inner", on = "country_code")

malnut_data.head()
```

```python
#Aggregating our results by item and sum of export quantities, and selecting only 25 top items
malnut_exports = malnut_data.groupby(["item_code","item"]).sum().reset_index()[["item_code","export_quantity"]].sort_values(by='export_quantity', ascending=False).head(25)

#Finding all data on those 25 items most exported by the countries with more than 10% undernourishment
data_25 = pd.merge(data_2018,malnut_exports[["item_code"]], how = "inner", on = "item_code")

#Creating a subset with only top 200 import quantities
top_200 = data_25.sort_values(by='import_quantity', ascending=False).head(200)

#Setting our boolean to true for this column
top_200["great_import_from_undern_countries"] = True

#Selecting only relevant columns before merging
top_200 = top_200[["country_code","item_code","great_import_from_undern_countries"]]

#Creating our final dataset with the relevant columns by adding a great_import_from_undern_countries set to true only for the
#right columns
data_final = pd.merge(data_2018,top_200, how="left",on=["country_code","item_code"])

#Setting boolean values for other cells to false
data_final.great_import_from_undern_countries = np.where(data_final.great_import_from_undern_countries.isna(),False, True)

#Verifying that we only have 200 columns set to true
data_final[(data_final.great_import_from_undern_countries == True)]
#That's a success ! That's the end of Task 4
```

### 5.1 Proportions of the global domestic supply used as food/feed/losses/other uses

```python
#We group by origin to separate vegetal and animal food
summary = data_final.groupby(["origin"]).sum().reset_index()

#We calculate the percentages here 
summary["food_pct"] = round((100*summary["food"] / summary["domestic_supply_quantity"]),2).astype(str) + '%'
summary["feed_pct"] = round((100*summary["feed"] / summary["domestic_supply_quantity"]),2).astype(str) + '%'
summary["losses_pct"] = round((100*summary["losses"] / summary["domestic_supply_quantity"]),2).astype(str) + '%'
summary["other_uses_pct"] = round((100*summary["other_uses"] / summary["domestic_supply_quantity"]),2).astype(str) + '%'

#Displaying our summary table
summary[["origin","domestic_supply_quantity","food_pct","feed_pct","losses_pct","other_uses_pct"]]
```

We can see that only 43% of the vegetal supply is used as food and 76% of the animal supply!

### 5.2 Number of humans which could be fed if plant-based food supply was used for human consumption?

```python
#Calculating the total number of available kcal and kg of food if feed was used for food
summary["feed_and_food_kcal"] = (summary.food_supply_kcal * summary.feed / summary.food) + summary.food_supply_kcal
summary["feed_and_food_kgprotein"] = (summary.food_supply_kgprotein * summary.feed / summary.food) + summary.food_supply_kgprotein

#Now let's find out how many humans this could feed
#The average diatery energy requirement in 2018 is 2359 (FAO DATA)
avg_kcal_req = 2359
summary["human_fed_kcal"] = summary.feed_and_food_kcal / avg_kcal_req / 365
#For protein, the average need is 50g per day or 0.05 kg
summary["human_fed_kgprotein"] = summary.feed_and_food_kgprotein / 0.05 / 365

#We take into account the world population in 2018 calculated in Task 2 :
summary["human_fed_pct_pop_kcal"] = 100*summary.human_fed_kcal / world_population
summary["human_fed_pct_pop_protein"] = 100*summary.human_fed_kgprotein / world_population

#Displaying our summary table
summary[["origin","human_fed_pct_pop_kcal","human_fed_pct_pop_protein"]]
```

We could feed 133% of the population by using only vegetal products !


### 5.3 Number of humans which could be fed with the global food supply?

```python
#Here we take into account the avg_kcal_req calculated earlier
human_fed_pct_kcal = 100*summary.sum().food_supply_kcal / avg_kcal_req / 365 / world_population
#For this part, we take into account the required protein per days which is 50g or 0.05 kg
human_fed_pct_kgprotein = 100*summary.sum().food_supply_kgprotein / 0.05 / 365 / world_population

[human_fed_pct_kcal, human_fed_pct_kgprotein]
```

We could feed 124% of the population with the current food supply, and fulfill the protein needs of 165% of the population

### Projected population size in 2050

```python
#Extracting historic world population data from the FAO website
world_pop = pd.read_csv("Data/world_population.csv")

#Keeping only relevant columns
world_pop = world_pop[["Year Code","Value"]]

#Renaming the columns
world_pop.columns = ["year","population"]

#Unit is 1000 persons, let's convert it into billions
world_pop["population"] = world_pop["population"] / 1000000

world_pop.head()
```

```python
from sklearn.linear_model import LinearRegression

A = world_pop.year.values.reshape(-1,1)
B = world_pop.population

lin_reg = LinearRegression()
lin_reg.fit(A,B)

years_extended = np.arange(1950, 2050, 1).reshape(-1,1)
plt.figure()
plt.title("World population evolution")
plt.scatter(A,B, label="Data")
plt.plot(years_extended,lin_reg.predict(years_extended),color="red",label="Linear Regression",linewidth=2)
plt.xlabel("Year")
plt.ylabel("Total World Population (billions)")
plt.legend()
plt.show()
```

```python
#Finding value in 2050
y_new = lin_reg.predict(np.arange(2050,2051,1).reshape(-1,1))

y_new[0]
```
